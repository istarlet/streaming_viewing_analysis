## Imdb Data

### 1. Read the Data In
This is where I import any needed Python libraries and all datasets I'll use in this notebook.

```python
# Import needed libraries
import pandas as pd
```

```python
# Read the CSV file in
imdb = pd.read_csv("datasets/titles.csv")
```

```python
# View the first five rows of data
imdb.head()
```

### 2. Manipulate and Clean the Data

```python
# Get a list of all the column names
imdb.columns
```

```python
# Drop the columns that aren't needed
imdb = imdb.drop(columns=["id", "description", "production_countries", "seasons", "imdb_id", "age_certification", "imdb_votes", "tmdb_score", "tmdb_popularity"], axis=1)

# View updated dataframe to make sure that columns were dropped
imdb.head()
```

```python
# Fix the column names in the dataframe
fixed_columns = {
    "title":"Title",
    "type":"Type",
    "release_year":"Release Year",
    "runtime":"Runtime",
    "genres":"Genres",
    "imdb_score":"Imdb Score"
}

# Check that the column names are displaying correctly
imdb.rename(columns=fixed_columns, inplace=True)
imdb.head()
```

```python
# Store the cleaned dataframe as a variable to use in the main notebook
imdb_cleaned = imdb
%store imdb_cleaned
```
